# Numpy First Program

```python
import numpy as np


arr = np.array([1, 2, 3, 4])
print(arr)
print(arr + 5)
print(arr * 8)
print(arr.mean())
print(arr.sum())
```

```python
arr = np.array([23, 64, 598, 84, 87, 45])
print(arr)
print(arr + 5)
print(arr * 8)
print(arr.mean())
print(arr.sum())
```
